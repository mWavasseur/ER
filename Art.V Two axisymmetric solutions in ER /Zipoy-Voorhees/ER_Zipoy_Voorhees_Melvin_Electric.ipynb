# ZIPOY-VOORHEES SOLUTION WITH AN EXTERNAL ELECTRIC FIELD IN ER

The dilaton-equivalent form of the action in ER reads:

\begin{eqnarray}
S
&=& \frac{1}{\tilde{\kappa}}\left(
\frac{\vartheta^{2} R}{2\tilde{\kappa}}
+ \vartheta \mathcal{L}_{m}\right).
\end{eqnarray}

The corresponding field equations are:

\begin{eqnarray}
\nabla_{\mu}\!\left(\vartheta F^{\mu\nu}\right) &=& 0, \\
3\vartheta^{-2}\Box\vartheta^{2}&=&\vartheta^{-1}\left(T-\mathcal{L}_{m}\right), \\
G_{\alpha\beta}
&=& \frac{T_{\alpha\beta}}{\vartheta}
+ \vartheta^{-2}\bigl(\nabla_{\alpha}\nabla_{\beta}
- g_{\alpha\beta}\Box \bigr)\vartheta^{2},
\end{eqnarray}

whith $\vartheta$ a scalar defined upon the ratio between the scalars $\mathcal{L}_{m}$ and $R$

\begin{equation}
    \vartheta := -\frac{\mathcal{L}_{m}}{R},
\end{equation}

and the stress-energy tensor 

\begin{equation}
T_{\mu\nu}=2\left(F_{\rho\mu}F^{\rho}_{\ \nu} - \frac{1}{4}g_{\mu\nu}F^{2}\right).
\end{equation}

We want to check if the Zipoy-Voorhees-Melvin solution with an electric field is solution of the ER equations. We refer the reader to the notebook EINSTEIN_Zipoy_Voorhees_Melvin for a complete description of the Zipoy Voorhees solution.

```python
functions_list=['Omega','Sigma','Lambda','Delta']
def subs_func(arg):
    subs_funcs = [(Lambda,Lambda_expr),(Sigma,Sigma_expr),(Delta,Delta_expr)] 
    if hasattr(arg, 'expr'):
        arg = arg.expr()
    if hasattr(arg, 'apply_map')*hasattr(arg, 'display'):
        for i, (old_func, new_func) in enumerate(subs_funcs):
            arg.apply_map(lambda f: f.substitute_function(old_func, new_func).subs(alpha=1/(2*(3)**(1/2))))
            print('{0} substituted'.format(functions_list[i])) 
    elif hasattr(arg, 'apply_map'):
        return arg.apply_map(lambda f: subs_func(f))
    else:
        for i, (old_func, new_func) in enumerate(subs_funcs):
            arg = arg.substitute_function(old_func, new_func).subs(alpha=1/(2*(3)**(1/2)))
            print('{0} substituted'.format(functions_list[i])) 
    print('Full Substitution : Done')
    return arg
```

```python
M = Manifold(4, 'M', structure='Lorentzian')
print(M)
```

```python
XY.<t,r,th,ph> = M.chart(r"t r:(0,+oo) th:(0,pi):\theta ph:(0,2*pi):\varphi")
XY
```

# I. Definition of the metric

```python
Delta = function('Delta')
Lambda = function('Lambda')
Sigma = function('Sigma')
Omega = function('Omega')
```

```python
delta, m, B, alpha= var('delta m B alpha')
alpha = 1/(2*sqrt(3))
assume(m >= 0, 'real')
assume(delta, 'real')
```

```python
g = M.metric()

g[0,0] = -Delta(r)**delta * Lambda(r,th)**((4*alpha**2+2)/(1+alpha**(2)))
g[1,1] = Delta(r)**(delta**2-delta-1)*Sigma(r,th)**(1-delta**2) * Lambda(r,th)**((4*alpha**2+2)/(1+alpha**(2)))
g[2,2] = Delta(r)**(delta**2-delta)*Sigma(r,th)**(1-delta**2) * Lambda(r,th)**((4*alpha**2+2)/(1+alpha**(2)))*r**2
g[3,3] = Delta(r)**(1-delta)*r**2*sin(th)**2 / Lambda(r,th)**((-4*alpha**2+2)/(1+alpha**(2)))

show(g.display())
show(LatexExpr(r'\sqrt{-g} = '),-(g.det().expr())^(1/2))
```

```python
Delta_expr(r) = 1 - 2 * m / r
Sigma_expr(r,th) = 1 - 2 * m /r + (m/r)**2 * (sin(th))**2
Lambda_expr(r,th) = 1 + (1+ alpha**2) / 4 * B**2 * Delta(r)**(1-delta) * (r*sin(th))**2

show(LatexExpr(r'\Delta = '), Delta_expr(r))
show(LatexExpr(r'\Sigma = '), Sigma_expr(r,th))
show(LatexExpr(r'\Lambda = '), Lambda_expr(r,th))
```

```python
g_nod = g.copy()
g_nod = subs_func(g_nod)
nod_ratio = (g_nod[3,3]/(sin(th)**2*g_nod[2,2])).factor()
show(LatexExpr(r"\lim_{\theta\rightarrow 0} \frac{g_{\phi\phi}}{\sin^{2}\theta g_{\theta\theta}} = " + latex(limit(nod_ratio.expr(), th=0).canonicalize_radical())))
```

```python
nab = g.connection(name=r'\nabla') 
```

# II. The vector potential

```python
pot_vec = M.tensor_field(0,1,name='A')
pot_vec[0]=-(delta*m+m-r)*B*cos(th)
pot_vec[1]=0
pot_vec[2]=0
pot_vec[3]=0

show(LatexExpr(r"A_{\mu\nu} = 0 \rightarrow A'_{\mu\nu} ="), pot_vec.display())
```

```python
DF = nab(pot_vec) ; DF
```

# III. Definition of the EM tensor

```python
F = diff(pot_vec)
F.set_name('F')
Fuu = F.up(g)

F.display()
```

# IV. Variables and functions

```python
Phi = M.scalar_field({XY: - alpha / (1+alpha**2) * log(Lambda(r,th))}, name='Phi', latex_name=r'\Phi')
Psi = M.scalar_field({XY: (Lambda(r,th))**(-2 * alpha **2 / (1+alpha **2 ))}, name='Psi') ; Psi.display()

Phi_elec = M.scalar_field({XY: -Phi.expr()}, name=r'\phi') 
Psi_elec = M.scalar_field({XY: 1/Psi.expr()}, name=r'\Psi')

varth_elec = M.scalar_field({XY:Psi_elec.expr()**(-1)}, name=r'\vartheta') # varth_elec=1/Psi_elec=e^{-2*\alpha*\phi}

show(LatexExpr(r"\Psi' = e^{2\alpha\phi'} = "), Psi_elec.expr())
show(LatexExpr(r"\varphi' = "), Phi_elec.expr())
show(LatexExpr(r"\vartheta = e^{-2\alpha\phi'} = \frac{1}{\Psi'} ="), varth_elec.expr())
```

# V. Field equations

```python
Ric = g.ricci() #Ricci tensor
```

```python
R = g.ricci_scalar() #Ricci scalar
```

```python
G = Ric-R*g/2 #Einstein tensor
```

## V.1. The scalar-field form $-\frac{\mathcal{L}_m}{R} = \vartheta$

```python
Lm = -F['_ij']*F.up(g)['^ij']/2
```

```python
eq = Lm/R + varth_elec
eq = subs_func(eq)
```

```python
latex_str = r'\frac{\mathcal{L}_m}{R} + \vartheta = ' 
show(LatexExpr(latex_str), numerator(eq.factor()).subs(abs(r-2*m)==r-2*m).subs(abs(m**2*sin(th)**2-2*m*r+r**2)==m**2*sin(th)**2-2*m*r+r**2).simplify_full())
```

```python
Lm_new_expr = -(B^2 * (r*(2*m - r) + (2*delta*m*r - (delta + 1)^2 * m^2) * sin(th)^2)/ (m^2 * sin(th)^2 - 2*m*r + r^2)) * ((m^2 * sin(th)^2 - 2*m*r + r^2) / (r*(r - 2*m)))^(delta^2) * (1 + (13/48)*B^2 * r^2 * sin(th)^2 * ((r - 2*m)/r)^(1 - delta))^(-56/13)
show(LatexExpr(r'\mathcal{L}_{m} = '), Lm_new_expr)
(subs_func(Lm).simplify_full()-Lm_new_expr).canonicalize_radical().is_zero()
```

```python
R_new_expr = (B^2 * (r*(2*m - r) + (2*delta*m*r - (delta + 1)^2 * m^2) * sin(th)^2)/ (m^2 * sin(th)^2 - 2*m*r + r^2)) * ((m^2 * sin(th)^2 - 2*m*r + r^2) / (r*(r - 2*m)))^(delta^2) * (1 + (13/48)*B^2 * r^2 * sin(th)^2 * ((r - 2*m)/r)^(1 - delta))^(-54/13)
show(LatexExpr(r'R = '), R_new_expr)
(subs_func(R).simplify_full()-R_new_expr).canonicalize_radical().is_zero()
```
