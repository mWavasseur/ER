# ZIPOY-VOORHEES SOLUTION IN EINSTEIN-MAXWELL-DILATON THEORIES

We consider external vacuum solutions in scalarÔÇôtensor
theories for the general class of actions

\begin{equation}
S = \int d^4x \sqrt{-g} \left[ R - 2 (\nabla \varphi)^2 - e^{-2\alpha\varphi}F^{2} \right],
\end{equation}

With the corresponding motion equations:

\begin{align*}
\nabla_\mu \left( e^{-2\alpha\phi} F^{\mu\nu} \right) = 0, \\
\nabla^2 \phi + \frac{a}{2} e^{-2\alpha\phi} F^2 = 0, \\
R_{\mu\nu} = 2 \nabla_\mu \phi \nabla_\nu \phi + 2 e^{-2\alpha\phi} F_{\mu\rho} F_\nu^{\ \rho} - \frac{1}{2} g_{\mu\nu} e^{-2\alpha\phi} F^2.
\end{align*}

We want to check if the Zipoy-Voorhees-Melvin solution is solution of the EMD equations:

\begin{equation}
ds^{2}=\Lambda^{\frac{2}{1+a^2}}
  \Bigl(
    \Delta^{\delta}\, dt^{2}
    - \Delta^{1-\delta}
      \Bigl[
        \Sigma^{1-\delta^{2}} \Delta^{\delta^{2}-1}
        \bigl( \tfrac{dr^{2}}{\Delta} + r^{2}\, d\theta^{2} \bigr)
      \Bigr]
  \Bigr)
  - \Delta^{1-\delta}\,\Lambda^{-\frac{2}{1+a^2}}\, r^{2}\sin^{2}\theta\, d\phi^{2}.
\end{equation}

\begin{equation}
\Delta = 1 - \frac{2\sigma}{r},
\qquad
\Sigma = 1 - \frac{2\sigma}{r} + \frac{\sigma^{2}}{r^{2}}\sin^{2}\theta.
\end{equation}



where  $\Lambda=1+\frac{1+a^2}{4} B_0^2 \rho^2$, with $e^{-2 a \Phi} = \Lambda^{\frac{2 a^2}{1+a^2}}$, and $A_\varphi = \frac{B_0 \rho^2}{2 \Lambda}$.

One has $\Psi = \Lambda^{-\frac{2 a^2}{1+a^2}}$ and $\Phi = - \frac{a}{1+a^2} \ln \Lambda$

```python
functions_list=['Sigma','Lambda','Delta']
def subs_func(arg):
    subs_funcs = [(Lambda,Lambda_expr),(Sigma,Sigma_expr),(Delta,Delta_expr)] 
    if hasattr(arg, 'expr'):
        arg = arg.expr()
    if hasattr(arg, 'apply_map')*hasattr(arg, 'display'):
        for i, (old_func, new_func) in enumerate(subs_funcs):
            arg.apply_map(lambda f: f.substitute_function(old_func, new_func))
            print('{0} substituted'.format(functions_list[i])) 
    elif hasattr(arg, 'apply_map'):
        return arg.apply_map(lambda f: subs_func(f))
    else:
        for i, (old_func, new_func) in enumerate(subs_funcs):
            arg = arg.substitute_function(old_func, new_func)
            print('{0} substituted'.format(functions_list[i])) 
    print('Full Substitution : Done')
    return arg
```

```python
M = Manifold(4, 'M', structure='Lorentzian')
print(M)
```

```python
XY.<t,r,th,ph> = M.chart(r"t r:(0,+oo) th:(0,pi):\theta ph:(0,2*pi):\varphi")
XY
```

# I. Definition of the metric

```python
Delta = function('Delta')
Lambda = function('Lambda')
Sigma = function('Sigma')
```

```python
delta, m, B, alpha= var('delta m B alpha')
assume(m >= 0, 'real')
assume(delta, 'real')
```

```python
g = M.metric()

g[0,0] = -Delta(r)**delta * Lambda(r,th)**(2/(1+alpha**(2)))
g[1,1] = Delta(r)**(delta**2-delta-1)*Sigma(r,th)**(1-delta**2) * Lambda(r,th)**(2/(1+alpha**(2)))
g[2,2] = Delta(r)**(delta**2-delta)*Sigma(r,th)**(1-delta**2) * Lambda(r,th)**(2/(1+alpha**(2)))*r**2
g[3,3] = Delta(r)**(1-delta)*r**2*sin(th)**2 / Lambda(r,th)**(2/(1+alpha**(2)))

show(LatexExpr(r'g_{\mu\nu} = '), g.display())
show(LatexExpr(r'\sqrt{-g} = '),-(g.det().expr())^(1/2))
```

```python
Delta_expr(r) = 1 - 2 * m / r
Sigma_expr(r,th) = 1 - 2 * m /r + (m/r)**2 * (sin(th))**2
Lambda_expr(r, th) = 1 + (1+ alpha**2) / 4 * B**2 * Delta(r)**(1-delta) * (r*sin(th))**2

show(LatexExpr(r'\Delta = '), Delta_expr(r))
show(LatexExpr(r'\Sigma = '), Sigma_expr(r,th))
show(LatexExpr(r'\Lambda = '), Lambda_expr(r,th))
```

```python
g_nod = g.copy()
g_nod = subs_func(g_nod)
nod_ratio = (g_nod[3,3]/(sin(th)**2*g_nod[2,2])).factor()
show(LatexExpr(r"\lim_{\theta\rightarrow 0} \frac{g_{\phi\phi}}{\sin^{2}\theta g_{\theta\theta}} = " + latex(limit(nod_ratio.expr(), th=0).canonicalize_radical())))
```

```python
nab = g.connection(name=r'\nabla') 
```

```python
Ric = g.ricci() #Ricci tensor
```

# II. The vector potential

```python
pot_vec = M.tensor_field(0,1,name='A')
pot_vec[0]=0
pot_vec[1]=0
pot_vec[2]=0
pot_vec[3] = 2 / (1+alpha**2) / B / Lambda(r,th)

show(LatexExpr(r"A_{\mu\nu} = 0 \rightarrow A'_{\mu\nu} ="), pot_vec.display())
```

```python
DF = nab(pot_vec) ; DF
```

# III. Definition of the EM tensor

```python
F = diff(pot_vec)
F.set_name('F')
Fuu = F.up(g)
F.display()
```

```python
epsilon_3D = M.tensor_field(0,3, antisymmetric=True, name=r'\epsilon_{c}') 
epsilon_3D[1,2,3] = 1
epsilon_3D[2,3,1] = 1
epsilon_3D[3,1,2] = 1
epsilon_3D[3,2,1] = -1
epsilon_3D[2,1,3] = -1
epsilon_3D[1,3,2] = -1
epsilon_3D.display()
```

```python
B_mag = M.tensor_field(1,0)  
for i in [1, 2, 3]: 
    expr = 1/2 * sum(epsilon_3D[i,j,k] * F[j,k] for j in [1,2,3] for k in [1,2,3])
    B_mag[i] = expr.expr().factor() 
show(LatexExpr(r'B^{\mu} = '),B_mag[:])
```

# IV. The field equations

```python
Psi = M.scalar_field({XY: (Lambda(r,th))**(-2 * alpha **2 / (1+alpha **2 ))}, name='Psi') ; Psi.display()
```

```python
Phi = M.scalar_field({XY: - alpha / (1+alpha**2) * log(Lambda(r,th))}, name='Phi', latex_name=r'\Phi')
Phi.display()
```

## IV.1. Verification of the Maxwell equations $\bigtriangledown_{\mu}(e^{-2\alpha\Phi}F^{\mu\nu})=0$

```python
eq1 = nab(Fuu / Psi)
eq1 = subs_func(eq1['^a._a'])
```

```python
show('Verification: ', LatexExpr(r'\bigtriangledown_{\mu}(e^{-2\alpha\Phi}F^{\mu\nu})=0: '), eq1.is_zero())
```

## IV.2. Verification of the equation $\bigtriangledown^{2}\Phi+\frac{\alpha}{2}e^{-2\alpha\Phi}F^{2}=0$ 

```python
F2 = F['_ab']*Fuu['^ab']
```

```python
eq2_1 = Phi.dalembertian()
```

```python
eq2_2 = (alpha / 2) * F2 / Psi 
```

```python
eq2 = eq2_1 + eq2_2
```

```python
F2 = F['_ab']*Fuu['^ab']
eq2_1 = Phi.dalembertian()
eq2_2 = (alpha/2)*F2 / Psi 
eq2 = eq2_1 + eq2_2
```

```python
eq_num = numerator(subs_func(eq2))
eq_num = eq_num.factor()
```

```python
LateX = LatexExpr(r'\bigtriangledown^{2}\Phi+\frac{\alpha}{2}e^{-2\alpha\Phi}F^{2}=0 ')
show('Verification: ', LateX, LatexExpr(r'\quad'), eq_num.canonicalize_radical().factor().is_zero())
```

## IV. 3. Verification of the main motion equation $R_{\mu\nu}=2\triangledown_{\mu}\Phi\triangledown_{\nu}\Phi+\frac{T_{\mu\nu}}{\Psi}$

```python
Fud = F.up(g,0)
T = 2*(F['_k.']*Fud['^k_.'] - 1/4*F2*g)
```

```python
UU=T / Psi
```

```python
nab_phi = nab(Phi)
S = 2*nab_phi*nab_phi     
```

```python
RHS = UU + S
```

```python
eq3 = Ric - RHS
eq3 = subs_func(eq3)
```

```python
LateX = LatexExpr(r'R_{\mu\nu}=2\triangledown_{\mu}\Phi\triangledown_{\nu}\Phi+\frac{T_{\mu\nu}}{\Psi} ')
show('Verification: ',LateX, LatexExpr(r'\quad'), eq3.is_zero())
```
